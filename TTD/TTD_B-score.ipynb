### Approve Flag

```python
def Bscore_Approve(x):
    if x=="C17" or x == "A":
        return "1"
    else :
        return "0"
    
def Pending(x):
    if x=="20" or x == "30"or x == "40"or x == "50"or x == "60"or x == "130":
        return "1"
    else :
        return "0"   

def Reject_by_cm(x):
    if x=="C17" :
        return "1"
    else :
        return "0" 

def Cancel_Flag():
    condition = [(df_main2['ขั้นตอน'] == 'ยกเลิกรายการระหว่างดำเนินการ')&(df_main2['Reject_by_cm'] == '1'),
                 ~((df_main2['ขั้นตอน'] == 'ยกเลิกรายการระหว่างดำเนินการ')&(df_main2['Reject_by_cm'] == '1'))]
    
    criterion = ['1','0']
    return np.select(condition, criterion)

def Score_pass():
    condition = [(df_main2['Bscore_Approve'] == '1')&(df_main2['Pending'] == '0'),
                 ~((df_main2['Bscore_Approve'] == '1')&(df_main2['Pending'] == '0'))]
    
    criterion = ['1','0']
    return np.select(condition, criterion)

def Suppose_Approve():
    condition = [(df_main2['Score_pass'] == '1')&(df_main2['Reject_by_cm'] == '0'),
                 ~((df_main2['Score_pass'] == '1')&(df_main2['Reject_by_cm'] == '0'))]
    
    criterion = ['1','0']
    return np.select(condition, criterion)

    

df_main2['Bscore_Approve'] = df_main2['app Sts'].apply(Bscore_Approve)
df_main2['Pending'] = df_main2['appstate'].apply(Pending)
df_main2['Reject_by_cm'] = df_main2['app Sts'].apply(Reject_by_cm)
df_main2['Cancel_Flag'] = Cancel_Flag()
df_main2['Score_pass'] = Score_pass()
df_main2['Suppose_Approve'] = Suppose_Approve()
```

### fail deviate

```python
def filter_deviate(df_Fail):
    #determine criteria
    con_hou_sec =           (   (df_Fail['รหัสประเภทผลิตภัณฑ์'] == 7100) &
                                (df_Fail['MOU_Flag'] == 'NonMOU') &    
                                (df_Fail['B-score level'] == 'DD') &
                                (df_Fail['B-score'] >= 690)
                            )

    con_mou_unsec_other =   (   (df_Fail['รหัสประเภทผลิตภัณฑ์'] == 7300) &
                                
                                (df_Fail['B-score level'] == 'EE') &
                                (df_Fail['B-score'] >= 673) &
                                (df_Fail['TypeMOU'] == 'Listed') 
                            )
    
    con_mou_unsec_listed =  (   (df_Fail['รหัสประเภทผลิตภัณฑ์'] == 7300) &
                                
                                (df_Fail['B-score level'] == 'EE') &
                                (df_Fail['B-score'] >= 673) &
                                (df_Fail['TypeMOU'] == 'Other') 
                            )   

    con_mou_unsec_gov =     (   (df_Fail['รหัสประเภทผลิตภัณฑ์'] == 7300) &
                                
                                (df_Fail['B-score level'] == 'EE') &
                                (df_Fail['TypeMOU'] == 'Gov/State') 
                            )
    
    con_mou_sec =           (   (df_Fail['รหัสประเภทผลิตภัณฑ์'] == 7200) &
                                (df_Fail['มีหลักประกัน/ไม่มี'] == 'มีหลักประกัน') &
                                (df_Fail['MOU_Flag'] == 'MOU') &
                                (df_Fail['B-score level'] == 'FF')  
                            )

    # filter data with criteria ----> get deviate no.application
    df_Fail = df_Fail[con_hou_sec | 
            con_mou_unsec_other | 
            con_mou_unsec_listed |
            con_mou_unsec_gov |
            con_mou_sec]

    return df_Fail

df_Fail_Deviate = filter_deviate(df_Fail)
df_Fail_Deviate  
```

### pass deviate

```python
def filter_deviate(df_pass):
    #determine criteria
    con_hou_sec =           (   (df_pass['รหัสประเภทผลิตภัณฑ์'] == 7100) &
                                (df_pass['MOU_Flag'] == 'NonMOU') &    
                                (df_pass['B-score level'] == 'DD') &
                                (df_pass['B-score'] >= 690)
                            )

    con_mou_unsec_other =   (   (df_pass['รหัสประเภทผลิตภัณฑ์'] == 7300) &
                                
                                (df_pass['B-score level'] == 'EE') &
                                (df_pass['B-score'] >= 673) &
                                (df_pass['TypeMOU'] == 'Listed') 
                            )
    
    con_mou_unsec_listed =  (   (df_pass['รหัสประเภทผลิตภัณฑ์'] == 7300) &
                                
                                (df_pass['B-score level'] == 'EE') &
                                (df_pass['B-score'] >= 673) &
                                (df_pass['TypeMOU'] == 'Other') 
                            )   

    con_mou_unsec_gov =     (   (df_pass['รหัสประเภทผลิตภัณฑ์'] == 7300) &
                                
                                (df_pass['B-score level'] == 'EE') &
                                (df_pass['TypeMOU'] == 'Gov/State') 
                            )
    
    con_mou_sec =           (   (df_pass['รหัสประเภทผลิตภัณฑ์'] == 7200) &
                                (df_pass['มีหลักประกัน/ไม่มี'] == 'มีหลักประกัน') &
                                (df_pass['MOU_Flag'] == 'MOU') &
                                (df_pass['B-score level'] == 'FF')  
                            )

    # filter data with criteria ----> get deviate no.application
    df_pass = df_pass[con_hou_sec | 
            con_mou_unsec_other | 
            con_mou_unsec_listed |
            con_mou_unsec_gov |
            con_mou_sec]

    return df_pass

df_Pass_Deviate = filter_deviate(df_pass)
df_Pass_Deviate  
```

```python
def ProbabilityBad():
    condition = [(df_all['ความน่าจะเป็นในการชำระหนี้คืน'] < 5),
                 (df_all['ความน่าจะเป็นในการชำระหนี้คืน'] >=5 )&(df_all['ความน่าจะเป็นในการชำระหนี้คืน'] <10),
                 (df_all['ความน่าจะเป็นในการชำระหนี้คืน'] >=10 )&(df_all['ความน่าจะเป็นในการชำระหนี้คืน'] <20),
                 (df_all['ความน่าจะเป็นในการชำระหนี้คืน'] >=20 )&(df_all['ความน่าจะเป็นในการชำระหนี้คืน'] <30),
                 (df_all['ความน่าจะเป็นในการชำระหนี้คืน'] >=30)
                 ]
    
    criterion = ['A:<5','B:5-20','C:10-<20','D:20-<30','E: 30+']
    return np.select(condition, criterion)
```
